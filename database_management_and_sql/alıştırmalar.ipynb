## 15.Çalışanları (employee) isimleri ile birlikte yaptıkları toplam satışlara göre sıralayınız. 

```python
%%sql
SELECT * FROM Employees
```

```python
%%sql
SELECT em.FirstName AS "İsim", em.LastName AS "Soyisim", price
FROM Orders o 
LEFT JOIN OrderDetails od ON o.OrderID = od.OrderID
LEFT JOIN Products pr ON od.ProductID = pr.ProductID
LEFT JOIN Employees em ON em.EmployeeID = o.EmployeeID
GROUP BY em.EmployeeID
ORDER BY SUM(pr.Price) DESC;
```

```python
%%sql
SELECT em.FirstName AS "İsim", em.LastName AS "Soyisim", SUM(pr.Price)
FROM Employees em 
LEFT JOIN Orders o ON em.EmployeeID = o.EmployeeID
LEFT JOIN OrderDetails od ON o.OrderID = od.OrderID
LEFT JOIN Products pr ON od.ProductID = pr.ProductID
GROUP BY em.EmployeeID
ORDER BY SUM(pr.Price) DESC;
```

```python
%%sql
SELECT em.FirstName, SUM(p.Price) toplam_satis 
FROM Orders o 
LEFT JOIN OrderDetails od on o.OrderID = od.OrderID 
LEFT JOIN Products p on od.ProductID = p.ProductID 
LEFT JOIN Employees em on o.EmployeeID = em.EmployeeID 
GROUP BY o.EmployeeID 
ORDER BY toplam_satis DESC;
```

## 16. Germany'dan verilen siparişlerin kategorilerine göre ortalama fiyatları nelerdir?

```python
%%sql
SELECT name, sql FROM sqlite_master
```

```python
%%sql
SELECT ca.CategoryName, AVG(pr.Price) FROM Categories ca 
LEFT JOIN Products pr on ca.CategoryID = pr.CategoryID
LEFT JOIN OrderDetails od on pr.ProductID = od.ProductID
LEFT JOIN Orders o on od.OrderID = o.OrderID
LEFT JOIN Customers cu on cu.CustomerID = o.CustomerID
WHERE cu.Country = "Germany"
GROUP BY ca.CategoryName
ORDER BY AVG(pr.Price) DESC;
```

## 17.Almanya ya da USA'den verilen siparişlerin kategorilerine göre ortalama fiyatları nelerdir?

```python
%%sql
SELECT ca.CategoryName, AVG(pr.Price) FROM Categories ca 
LEFT JOIN Products pr on ca.CategoryID = pr.CategoryID
LEFT JOIN OrderDetails od on pr.ProductID = od.ProductID
LEFT JOIN Orders o on od.OrderID = o.OrderID
LEFT JOIN Customers cu on cu.CustomerID = o.CustomerID
WHERE cu.Country = "Germany" OR cu.Country = "USA"
GROUP BY ca.CategoryName
ORDER BY AVG(pr.Price) DESC;
```

## 19.Müşterilerin 1997 yılına ait siparişlerinin maksimum miktarlarını bulup müşterilerin isimleri ile birlikte büyükten küçüğe sıralayınız.

```python
%%sql
SELECT o.OrderDate,MAX(pr.Price) maks,cu.CustomerName FROM Categories ca 
LEFT JOIN Products pr on ca.CategoryID = pr.CategoryID
LEFT JOIN OrderDetails od on pr.ProductID = od.ProductID
LEFT JOIN Orders o on od.OrderID = o.OrderID
LEFT JOIN Customers cu on cu.CustomerID = o.CustomerID
WHERE o.OrderDate LIKE "%1997"
GROUP BY cu.CustomerName
ORDER BY maks DESC;
```

```python
%%sql
SELECT CustomerName, MAX(Price) maks 
FROM Products 
LEFT JOIN OrderDetails on Products.ProductID = OrderDetails.ProductID 
LEFT JOIN Orders on OrderDetails.OrderID = Orders.OrderID 
LEFT JOIN Customers on Customers.CustomerID = Orders.CustomerID 
WHERE OrderDate like '%1997%' 
GROUP BY CustomerName ORDER BY maks DESC;
```
